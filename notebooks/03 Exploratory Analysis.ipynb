### Distribution Differences

See the difference in distribution between sentences containing specified tags (generated in the feature creation step) and the underlying sentiment distribution.

```python
def get_dist_diffs(df_phrase, tag):
    '''Helper function for plots. Returns difference in underlying distribution and distribution for input tag.'''
    # get underlying distribution summary
    df_summary = df_phrase.groupby('sentiment').count().reset_index()[['sentiment','text']]
    df_summary.text = df_summary.text.apply(lambda x: x / df_summary.text.sum())
    df_summary.columns = ['sentiment', 'distribution']
    
    # get distribution for tag summary
    df_summary_tag = df_phrase[df_phrase[tag]!=0]
    df_summary_tag = df_summary_tag.groupby('sentiment').count().reset_index()[['sentiment','text']]
    df_summary_tag.text = df_summary_tag.text.apply(lambda x: x / df_summary_tag.text.sum())
    df_summary_tag.columns = ['sentiment', 'tag_distribution']

    # get distribution difference
    df_comparison = pd.merge(df_summary, df_summary_tag, 'inner', left_on='sentiment', right_on='sentiment')
    df_comparison['distribution_diff'] = df_comparison.tag_distribution - df_comparison.distribution
    
    # return distributino difference
    return df_comparison
```

```python
# manual ordering for write up
tags = ['count_lagging:up', 'count_leading:up', 'count_pos', 'count_up', 'count_lagging:down',
        'count_leading:down', 'count_down','count_neg','count_lagging','count_leading']
```

```python
# create dataframe containing distribution difference 
for i, tag in enumerate(tags):
    if i==0:
        df_comparison_all = get_dist_diffs(df_phrase, tag)
        df_comparison_all['tag'] = tag
    
    else:
        df_comparison = get_dist_diffs(df_phrase, tag)
        df_comparison['tag'] = tag

        df_comparison_all = df_comparison_all.append(df_comparison)

# special case - no tags are present in the sentence 
df_comparison = df_phrase[df_phrase.tags=='']

df_summary = df_phrase.groupby('sentiment').count().reset_index()[['sentiment','text']]
df_summary.text = df_summary.text.apply(lambda x: x / df_summary.text.sum())
df_summary.columns = ['sentiment', 'distribution']

df_summary_tag = df_phrase[df_phrase.tags=='']
df_summary_tag = df_summary_tag.groupby('sentiment').count().reset_index()[['sentiment','text']]
df_summary_tag.text = df_summary_tag.text.apply(lambda x: x / df_summary_tag.text.sum())
df_summary_tag.columns = ['sentiment', 'tag_distribution']

df_comparison = pd.merge(df_summary, df_summary_tag, 'inner', left_on='sentiment', right_on='sentiment')
df_comparison['distribution_diff'] = df_comparison.tag_distribution - df_comparison.distribution
df_comparison['tag'] = 'no_tags_present'

df_comparison_all = df_comparison_all.append(df_comparison)
```

```python
# create facetgrid
g = sns.FacetGrid(df_comparison_all, col='tag', col_wrap=4);
g.map(sns.barplot, 'sentiment', 'distribution_diff', order=order, palette=palette);

g.fig.suptitle('Percentage difference between overall sentiment distribution and sentiment distribution of sentences containing given tags.', 
              y=1.05,
               size=12);

g.set_axis_labels("Sentiment", "% Difference")

from matplotlib.ticker import FuncFormatter
def to_percent(y, position):
    s = str(round(100*y))
    return s + '%'
formatter = FuncFormatter(to_percent)
plt.gca().yaxis.set_major_formatter(formatter)
plt.show()

#g.savefig("images/fig2_distribution_difference.png")
```
